##Introduction

Market Basket Analysis is a data mining technique used to discover relationships between items frequently purchased together. It helps businesses understand customer purchasing behavior and improve sales through cross-selling strategies.

This project uses the Apriori algorithm to identify frequent itemsets and generate association rules from transaction data.

##Objective
>To analyze customer transaction data

>To identify frequently purchased item combinations

>To generate association rules using Apriori algorithm

>To provide business insights for product bundling

###Import Libraries

```python
!pip install mlxtend

import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

###Dataset

We use a simple transactional dataset where each row represents a product purchased in a transaction.

```python
# Load dataset
df = pd.read_csv('groceries.csv', sep=';', header=None)

# Convert to list
df_list = df[0].str.split(',')

# Convert to basket
basket = pd.get_dummies(df_list.apply(pd.Series).stack()).groupby(level=0).sum()

# Convert to 0/1
basket = basket.applymap(lambda x: 1 if x > 0 else 0)

basket.head()
```

###Apply Apriori Algorithm

```python
frequent_items = apriori(basket, min_support=0.02, use_colnames=True)
frequent_items.head()
```

###Generate Association Rules

```python
rules = association_rules(frequent_items, metric="lift", min_threshold=1)
rules.head()
```

###Filter Strong Rules

```python
rules = rules[(rules['confidence'] > 0.3) & (rules['lift'] > 1)]
rules.head()
```

##Results Interpretation
Support shows how often items appear together
Confidence shows probability of buying related items
Lift shows strength of association

##Sample Insights
Customers who buy whole milk often buy other vegetables
Strong associations help in product bundling
Useful for marketing and store layout

##Conclusion

This project demonstrates how Market Basket Analysis can be used to identify relationships between products. The Apriori algorithm helps generate meaningful association rules that can improve business decision-making.

##Output

```python
rules = rules.sort_values(by='lift', ascending=False)
rules.head(10)
```

```python
rules_final = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
rules_final.head()
```

```python
rules_final['antecedents'] = rules_final['antecedents'].apply(lambda x: ', '.join(list(x)))
rules_final['consequents'] = rules_final['consequents'].apply(lambda x: ', '.join(list(x)))

rules_final.head()
```

```python
rules_final.to_csv('mba_results.csv', index=False)
```
